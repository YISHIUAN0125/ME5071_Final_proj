```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

yolo = YOLO('runs/segment/train/weights/best.pt').to('cuda')
img = cv2.imread('data/Cabbage.v1i.yolov8/valid/images/frame_00013_jpg.rf.f6f0c0562527360945ee8efe0d427712.jpg')
img = cv2.resize(img, (1000, 1000))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
results = yolo.predict(img, imgsz=1000, conf=0.1)
plt.imshow(results[0].plot())
plt.axis('off')
plt.show()
```
